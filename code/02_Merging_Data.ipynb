# 02 Merging Data

```python
import pandas as pd
```

```python
amer_df = pd.read_csv('../data/american_plots.csv', index_col=0)
amer_df.head()
```

```python
ind_df = pd.read_csv('../data/indian_plots.csv', index_col=0)
ind_df.head()
```

```python
ind_df['is_indian'] = 1
ind_df.head()
```

```python
amer_df['is_indian'] = 0
amer_df.head()
```

```python
df = pd.concat([amer_df, ind_df])
df
```

```python
df.to_csv('../data/merged_plots.csv')
```
